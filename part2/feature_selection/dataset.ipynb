```python
#%pip install pandas
#%pip install matplotlib

import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("dataset.txt", sep=None, engine="python")

# Basic overview
display(df.head())

# Plot settings
numeric_df = df.select_dtypes(include="number")

plt.figure(figsize=(8, 4))
plt.plot(df.index, numeric_df.iloc[:, 0])
plt.title("Line Plot of First Numeric Column")
plt.xlabel("Index")
plt.ylabel(str(numeric_df.columns[0]))
plt.tight_layout()
plt.show()

# Histogram for all numeric columns
df.select_dtypes(include="number").hist(figsize=(10, 6))
plt.tight_layout()
plt.show()
```
